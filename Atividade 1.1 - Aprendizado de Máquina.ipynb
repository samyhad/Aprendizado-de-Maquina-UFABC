```python
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
X, y = fetch_california_housing(return_X_y=True, as_frame=True)
a = 15
b = 15
plt.subplot(2, 4, 1)
X['MedInc'].plot.hist(figsize=(a,b), bins=10).legend(['MedInc'])
plt.subplot(2, 4, 2)
X['HouseAge'].plot.hist(figsize=(a,b), bins=10).legend(['HouseAge'])
plt.subplot(2, 4, 3)
X['AveRooms'].plot.hist(figsize=(a,b), bins=50).legend(['AveRooms'])
plt.subplot(2, 4, 4)
X['AveBedrms'].plot.hist(figsize=(a,b), bins=50).legend(['AveBedrms'])
plt.subplot(2, 4, 5)
X['Population'].plot.hist(figsize=(a,b), bins=30).legend(['Population'])
plt.subplot(2, 4, 6)
X['AveOccup'].plot.hist(figsize=(a,b), bins=100).legend(['AveOccup'])
plt.subplot(2, 4, 7)
X['Longitude'].plot.hist(figsize=(a,b), bins=10).legend(['Longitude'])
plt.subplot(2, 4, 8)
X['Latitude'].plot.hist(figsize=(a,b), bins=10).legend(['Latitude'])



```
